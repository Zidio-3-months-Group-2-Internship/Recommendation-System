# Popularity Based Recommendation System

### Top 50 books with highest rating will be displayed
### Books with more than or equal to 250 books will be considered

```python
# Merging Books and Rating on "ISBN"
ratings_with_name = ratings.merge(books, on="ISBN")
```

```python
ratings_with_name.head()
```

```python
 ratings_with_name.shape
```

```python
num_rating_df = ratings_with_name.groupby("Book-Title").count()['Book-Rating'].reset_index(
)
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby("Book-Title")['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on="Book-Title")
popular_df
```

```python
# Selecting top 50 books which have bee rated by more than 250 people and sorting it by avg_rating
popular_df = popular_df[popular_df['num_ratings'] >=250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
# Retrieving info like Book-Title, Book-Author , Image-URL-S, num-ratings, avg-ratings
popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')
```

# Collaborative Filtering Based Recommendation System

```python
# selecting books on the basis of people who have rated by at least 200 users
x = ratings_with_name.groupby("User-ID").count()['Book-Rating'] > 200
```

```python
good_users = x[x].index
```

```python
# REtriving data on the basis of good users who have rated at least 200 books
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(good_users)]
```

```python
# Retrieving books with more than 50 in rating
y = filtered_rating.groupby("Book-Title").count()['Book-Rating'] >=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
# Creating this table with user in column, Book-Title in rows
pt = final_ratings.pivot_table(index = "Book-Title", columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

### Every book is represented as Vector with 810 cells. We will plot all these books in a plot of 810 dimensions. We will then calculate distance or angle to check the books which are similar

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
# Making a function  which will recommend us 5 books on the basis on the book we used
def recommend(book_name):
    # fetching index
    index = np.where(pt.index==book_name)[0][0]
    # festching the similar items of the book
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:6]

    for i in similar_items:
        print(pt.index[i[0]])
    
```

```python
recommend('4 Blondes')
```
